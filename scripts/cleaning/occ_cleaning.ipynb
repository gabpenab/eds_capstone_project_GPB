2. Migratory species of Leptonycteris and Lasiurius has already been done by other researchers


```r
#set directory
getwd()
setwd("/home/user/capstone_project/data/raw_data")
list.files()

#load libraries
library (dplyr)
library (ggplot2)
library (readr)

#load data and explore
natalus_occ <- read_tsv("natalus_ocurrence.csv")
#head(natalus_occ)
str(natalus_occ)
summary(natalus_occ)

#select variables
natalus_occ_v <- natalus_occ %>% 
                 select("locality","stateProvince", "occurrenceStatus","individualCount", "decimalLatitude", "decimalLongitude", "coordinateUncertaintyInMeters", "elevation",
                       "eventDate", "year","basisOfRecord","institutionCode", "issue")

#filter NA for coordinates
natalus_occ_f <- natalus_occ_v %>% 
                 filter (!is.na(decimalLatitude) & !is.na(decimalLongitude))
dim(natalus_occ_f)
str(natalus_occ_f)

# occurence by year and date 
natalus_year_region <- natalus_occ_f %>% 
                       group_by(year,stateProvince) %>% 
                       summarise(count = n(), .groups = "drop")
natalus_year_region
plot_natalus_yr <-ggplot (natalus_year_region, aes(x=year, y=count, fill= stateProvince)) +
                  geom_col () +
                  labs(title = "Occurence per year and region",
                   x = "year",
                   y = "Count",
                   fill = "State")
plot_natalus_yr
natalus_occ_f %>%
  count(locality)
```

3. Oilbirds dataset 6619 records

```r
#load data and explore
oilbird_occ <- read_tsv("oilbirds_occ.csv")
#head(natalus_occ)
str(oilbird_occ)
summary(oilbird_occ)

#select variables
oilbird_occ_v <- oilbird_occ %>% 
                 select("locality","stateProvince", "occurrenceStatus","individualCount", "decimalLatitude", "decimalLongitude", "coordinateUncertaintyInMeters", "elevation",
                       "eventDate", "year","basisOfRecord","institutionCode","collectionCode", "issue")

#filter NA for coordinates
oilbird_occ_f <- oilbird_occ_v %>% 
                 filter (!is.na(decimalLatitude) & !is.na(decimalLongitude))
dim(oilbird_occ_f)
str(oilbird_occ_f)

# occurence by year and date 
oilbird_year_region <- oilbird_occ_f %>% 
                       group_by(year,stateProvince) %>% 
                       summarise(count = n(), .groups = "drop")
oilbird_year_region
plot_oilbird_yr <-ggplot (oilbird_year_region, aes(x=year, y=count, fill= stateProvince)) +
                  geom_col () +
                  labs(title = "Occurence per year and region",
                   x = "year",
                   y = "Count",
                   fill = "State")
plot_oilbird_yr
oilbird_occ_f %>%
  count(locality)
```
